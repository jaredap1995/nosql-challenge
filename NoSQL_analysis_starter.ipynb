### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
unique_authorities = establishments.distinct('LocalAuthorityName')
indices = [i for i, s in enumerate(unique_authorities) if 'London' in s]
print(indices)
unique_authorities[14]
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

#Non-Regex
establishments_london = establishments.find({'$and': [{'LocalAuthorityName': unique_authorities[14]}, {'RatingValue': {'$gte': 4}}]})
#Regex
establishments_london = establishments.find({'$and': [{'LocalAuthorityName': {'$regex': 'London'}}, {'RatingValue': {'$gte': 4}}]})

# Convert the result to a Pandas DataFrame
df_london = pd.DataFrame(list(establishments_london))

# Use count_documents to display the number of documents in the result
#Non-regex
result = establishments.count_documents({'$and': [{'LocalAuthorityName': unique_authorities[14]}, {'RatingValue': {'$gte': 4}}]})
#Regex
result = establishments.count_documents({'$and': [{'LocalAuthorityName': {'$regex': 'London'}}, {'RatingValue': {'$gte': 4}}]})

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5


degree_search = 0.01
penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})
latitude = penang_flavours['geocode']['latitude']
latitude=51.49014200
longitude = penang_flavours['geocode']['longitude']
longitude=0.08384000 #Just copy pasting and hardcoding because this is annoying


query = establishments.find({
    '$and': [
        {'geocode.latitude': {'$gte': latitude - degree_search}},
        {'geocode.latitude': {'$lte': latitude + degree_search}},
        {'geocode.longitude': {'$gte': longitude - degree_search}},
        {'geocode.longitude': {'$lte': longitude + degree_search}},
        {'RatingValue': 5}
    ]
}).sort('scores.Hygiene', pymongo.ASCENDING).limit(5) #Ascending because a low hygeine score is better according to the bootcamp spot information, so if we want the lowest
                                                        #hygeine score first we want the highest number first

results = list(query)
```

```python
# Convert result to Pandas DataFrame
df_penang_flavours = pd.DataFrame(results)
df_penang_flavours
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {'$match': {'scores.Hygiene': 0}},
    {'$group': {'_id': '$LocalAuthorityName','count': {'$sum': 1 }}},
    {'$sort': {'count': -1}}]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result

# Print the first 10 results
print("Number of documents in the result: ", len(results))
print("First 10 results: ", results[:10])
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene_0 = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(len(df_hygiene_0))

# Display the first 10 rows of the DataFrame
df_hygiene_0.head(10)
```
